```python
total= pd.read_csv("Twitch_game_data.csv", sep=",", encoding="'cp1252'")
total=total.drop(["Peak_channels","Streamers","Avg_channels"], axis=1)
total
```

```python
#IMPORTAR EL DATASET Y RENOMBRAR COLUMNAS
total1=total.set_index(["Year"])
total1.rename({"Avg_viewer_ratio":"Avg_H.Watched/H.Streamed"}, axis=1, inplace=True)
total1=total1[["Month","Rank","Game","Hours_watched","Hours_Streamed","Peak_viewers","Avg_viewers","Avg_H.Watched/H.Streamed"]]
total1
```

```python
total1["Hours_watched"] = total1["Hours_watched"].str.split("(", expand=True)[0]
total1["Hours_Streamed"] = total1["Hours_Streamed"].str.split("(", expand=True)[0]
total1["Hours_Streamed"] = total1["Hours_Streamed"].str.split("h", expand=True)[0]
total1["Peak_viewers"] = total1["Peak_viewers"].str.split("(", expand=True)[0]
total1["Avg_viewers"] = total1["Avg_viewers"].str.split("(", expand=True)[0]
total1["Avg_H.Watched/H.Streamed"] = total1["Avg_H.Watched/H.Streamed"].str.split("(", expand=True)[0]
total1
```
